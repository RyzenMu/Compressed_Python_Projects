```python
# installing pandas for opening the csv file

import pandas as pd

# open the csv to a variable

schedule_data = pd.read_csv('SCHEDULER.csv')

# read the csv file using pandas
schedule_data
```

```python


# converting date into a week day

# getting date column

date_column = schedule_data['DATE']

date_column
```

```python
# storing the unique dates in a variable 
coding_dates = date_column.unique()

coding_dates
```

```python
# printing the array
for date in coding_dates:    
    # typecasting float type to string type
    date_str = str(date)
    print(date_str)
```

```python
# finding the type of the date element

print(type(date_str))
```
