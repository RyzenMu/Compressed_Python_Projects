```python
# installing pandas for opening the csv file

import pandas as pd

# open the csv to a variable

schedule_data = pd.read_csv('SCHEDULER.csv')

# read the csv file using pandas
schedule_data
```

```python


# converting date into a week day

# getting date column

date_column = schedule_data['DATE']

date_column
```

```python
# storing the unique dates in a variable 
coding_dates = date_column.unique()

coding_dates
```

```python



last = '05/09/2022'

day = {
0 : 'Sunday',
1 : 'Monday',
2 : 'Tuesday',
3 : 'Wednesday',
4 : 'Thrusday',
5 : 'Friday',
6 : 'Saturday',}

my_dict = {}

my_date = []

my_day = []

for i in range(7,10):
    v = -1
    for j in range(1, 32):
        if (j < 10 and i == 7) or (j > 4 and i == 9):
            continue 

        

        if (j < 10):
            v += 1
            print(f'0{j}/0{i}/2022 - {day[v%7]}')
            my_date.append(f'0{j}/0{i}/22')
            my_day.append(day[v%7])
           
        else:  
            v += 1  
            print(f'{j}/0{i}/2022 - {day[v%7]}')
            my_date.append(f'{j}/0{i}/22')
            my_day.append(day[v%7])
            

```

```python
for i in range(len(my_date)):
    my_dict[my_date[i]] = my_day[i]
```
