```python
# installing pandas for opening the csv file

import pandas as pd

import datetime

import calendar

from datetime import date

# open the csv to a variable

schedule_data = pd.read_csv('SCHEDULER.csv')

# read the csv file using pandas
schedule_data
```

```python


# converting date into a week day

# getting date column

date_column = schedule_data['DATE']

date_column
```

```python
# storing the unique dates in a variable 
coding_dates = date_column.unique()

coding_dates
```

```python
# Converting string/float into datetime object
coding_dates = pd.to_datetime(coding_dates, format="%d/%m/%y")
coding_dates
```

```python
# Assining days to the date
def findDay(date):
    day, month, year = (int(i) for i in date.split('/'))
    date = datetime.date(year, month, day)
    return date.strftime('%A')

my_date = '31/08/2022'

print(findDay(my_date))
```
