```python
# installing pandas for opening the csv file

import pandas as pd

# open the csv to a variable

schedule_data = pd.read_csv('SCHEDULER.csv')

# read the csv file using pandas
schedule_data
```

```python


# converting date into a week day

# getting date column

date_column = schedule_data['DATE']

print(date_column)

```

```python
duration_column = schedule_data['DURATION']
```

```python
duration_column[0]
```

```python
patt = []
```

```python

for letter in duration_column[0]:
    
    if letter == ':':
        continue
    else:        
        
        patt.append(letter)

print(patt)
```

```python
pattern = [patt[0]+patt[1], patt[2]+patt[3], patt[4]+patt[5]]
print(pattern)
```

```python
for i in range(len(duration_column)):

    patt = []

    for letter in duration_column[i]:
    
        if letter == ':':
            continue
        else:        
        
            patt.append(letter)

    

    pattern = [patt[0]+patt[1], patt[2]+patt[3], patt[4]+patt[5]]
    

    for j in range(3):
        pattern[j] = int(pattern[j])

    time_sumation = pattern[0]*3600 + pattern[1]*60 + pattern[2]
    print(i, date_column[i], time_sumation)
    

    
```
